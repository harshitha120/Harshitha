```
#Write a Python program to find the total number of times letter 'p' is appeared in the below string using loop:-
    
#Input:- 'peter piper picked a peck of pickled peppers.\n'
#Output:- 9
a='peter piper picked a peck of pickled peppers.\n'
b=0
for i in range(0,len(a)):
  if (a[i]=='p'):
    b=b+1
print("The total number of times the letter p has repeated is:",b)



```

```
#Write a python program to find below output using loop:-

#Input:- 'Peter Piper Picked A Peck Of Pickled Peppers.'
#Output:- 'Peter piper picked a peck of pickled peppers'
int_str='peter piper picked a peck of pickled peppers.'
out_str=''
out_list=[]
j=0
for i in int_str:
  if i==' ' or i=='.':
    out_list.append(out_str)
    out_str=''
  else:
    out_str=out_str+i

while j<len(out_list):
  if j==0:
    out_str+=out_list[j]+''
  else:  
    out_str=out_str+" "+out_list[j].lower()
  j=j+1
print(out_str)
```
