# Matplot function

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
x=np.arange(1,10)
y=np.arange(11,20)
#scatter plot
plt.scatter(x,y,c='r')
plt.xlabel('x data set')
plt.ylabel('y data set')
plt.title('graph1')
```

```python
#plot function:
y=(x*x*x)
plt.plot(x,y,'b*--',linewidth=3)
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('graph2')
```

```python
plt.bar(x,y)
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('graph3')
```

```python
#creating subplot
x=np.arange(1,7)
y=np.arange(11,17)
y1=-(x*9/67)
y2=(x*9/56)
plt.subplot(2,2,1)
plt.plot(x,y,c='r')
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('graph1')
plt.subplot(2,2,2)
plt.plot(x,y1,c='b')
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('graph2')
plt.subplot(2,2,3)
plt.plot(x,y2,c='g')
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('graph3')
```

```python
#Plotting linear regression diagrams
x=np.arange(11,17)
b0=12
b1=17
y=b0+x*b1
plt.plot(x,y,'b*--',c='r')
```

```python
#Generating sine waves:
pi=3.14
x=np.arange(0,3.5,0.5)
y=np.sin(pi*x)
plt.plot(y)
plt.xlabel('X data')
plt.ylabel('y data')
plt.title('graph4')
```

```python
#plotting sine and cos
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
y1=np.cos(x)
#plotting the graphs:sine graph
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('graph1')
plt.xlabel('x data')
plt.ylabel('y data')

#plotting the graphs:cos graph
plt.subplot(2,1,2)
plt.plot(x,y1,c='g')
plt.title('graph2')
plt.xlabel('x data')
plt.ylabel('y data')
```

```python
#Plotting the bar graphs:
x1=[3,6,7]
y1=[20,40,10]
x2=[2,5,9]
y2=[45,34,90]
plt.bar(x1,y1,color='r')
plt.bar(x2,y2,color='g')
```

```python
#Histograms:
a=np.array([22,32,45,67,89,22,34,34,34,56,78,90,65,23,22,22])
plt.hist(a,bins=30)
plt.title('graph2')
plt.xlabel('x data')
plt.ylabel('y data')
```

```python
#Box plot:
data=[np.random.normal(0,std,100) for std in range(1,5)]
plt.boxplot(data,vert=True,patch_artist=True)
```

```python
#piechart:
labels='maths','Science','Social','Kannada'
sizes=[215,130,245,210]
colors=['gold','yellow','red','blue']
explode=(0.4,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,shadow=True)
```

# Seaborn


```python
import seaborn as sns
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
#correlation-heatmap-helps us to see the correelation between two factors
#To find correlation
data.corr()
```

```python
sns.heatmap(data.corr())
```

```python
#Joint plot:
sns.jointplot(x='tip',y='total_bill',data=data,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=data,kind='kde')
```

```python
sns.jointplot(x='tip',y='total_bill',data=data,kind='reg')
```

```python
sns.jointplot(x='tip',y='total_bill',data=data,kind='scatter')
```

```python
#if we have more than two independent features:
#We use pair plot:
sns.pairplot(data,hue='sex')
```

```python
#Distplot:
sns.distplot(data['tip'],kde=False,bins=10)
```

```python
sns.distplot(data['tip'],kde=True,bins=10)
```

# Categorical Plots

```python
sns.countplot('day',data=data)
```

```python
sns.countplot(y='day',data=data) 
```

```python
#barplot
sns.barplot(y='total_bill',x='sex',data=data)
```

```python
#box plot
sns.boxplot(y='total_bill',x='sex',data=data)
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=data,palette='rainbow')
```

```python
#violin plot:violin plot combination of box plot and Kde
sns.violinplot(x='total_bill',y='sex',data=data)
```
